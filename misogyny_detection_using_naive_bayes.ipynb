```python
import pandas as pd
X_train = pd.read_csv("trac_train.csv")
X_test = pd.read_csv("trac_test.csv")
y_test= pd.read_csv("trac_test.csv")
y_train= pd.read_csv("trac_train.csv")
X_train.groupby('Sub-task B').describe()
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_train.drop(X_train.columns[[0,2,3]],axis=1,inplace=True);
X_train.head()
```

```python
X_test.drop(X_test.columns[[0,2,3]],axis=1,inplace=True);
X_test.head()
```

```python
y_train.drop(y_train.columns[[0,1,2,]],axis=1,inplace=True);
y_train.head()
```

```python
y_test.drop(y_test.columns[[0,1,2]],axis=1,inplace=True);
y_test.head()
```

```python
y_train['task']=y_train["Sub-task B"].apply(lambda x : 1 if(x=='GEN') else 0)
y_train.head()
```

```python
y_test['task']=y_test["Sub-task B"].apply(lambda x : 1 if(x=='GEN') else 0)
y_test.head()
```

```python
y_train.drop(['Sub-task B'],axis=1,inplace=True)
y_train.head()
```

```python
y_test.drop(['Sub-task B'],axis=1,inplace=True)
y_test.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer(min_df=0, lowercase=False)
X_train_count = v.fit_transform(X_train.Text)
X_train_count.toarray()[:3]
```

```python
v.vocabulary_
```

```python
sentences = ['Rashmi likes ice cream', 'Rashmi hates chocolate.']

vectorizer = CountVectorizer()
vectorizer.fit(sentences)
vectorizer.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train.task)
```

```python
emails = [
    'Hey mohan can we get together to watch footbal game tommorow?',
    'Upto 20% discount on parking exclusive offer just for you. Dont miss this reward'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test.Text)
X_test_count.shape
```

```python
model.score(X_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train.Text,y_train.task)
```

```python
clf.score(X_test.Text,y_test)
```
